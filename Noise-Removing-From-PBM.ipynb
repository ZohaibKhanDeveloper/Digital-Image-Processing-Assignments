```python
import cv2
import numpy as np
```

```python
def remove_noise_hmt(img,img_num):
    kernel1 = np.array([[0, 0, 0],
                         [0, 1, 0],
                         [0, 0, 0]], dtype=np.uint8)
    kernel2 = np.array([[1, 1, 1],
                         [1, 0, 1],
                         [1, 1, 1]], dtype=np.uint8)
    output1 = cv2.morphologyEx(img, cv2.MORPH_HITMISS, kernel1)
    output2 = cv2.morphologyEx(img, cv2.MORPH_HITMISS, kernel2)
    filtered_img = img - (output1 + output2)
    cv2.imwrite(f'PBM-Images/filtered_image_hmt{img_num}.pbm', filtered_img)
```

```python
img1 = cv2.imread("PBM-Images/test1.pbm",0)
img2 = cv2.imread("PBM-Images/testChars.pbm",0)
img3 = cv2.imread("PBM-Images/testHard.pbm",0)
img4 = cv2.imread("PBM-Images/testPacmen.pbm",0)
img5 = cv2.imread("PBM-Images/testPacmenSubset.pbm",0)
img6 = cv2.imread("PBM-Images/testSimple.pbm",0)
```

```python
img1 = ~ img1
img2 = ~ img2
img3 = ~ img3
img4 = ~ img4
img5 = ~ img5
img6 = ~ img6
```

```python
remove_noise_hmt(img1,1)
remove_noise_hmt(img2,2)
remove_noise_hmt(img3,3)
remove_noise_hmt(img4,4)
remove_noise_hmt(img5,5)
remove_noise_hmt(img6,6)
```
