```python
import cv2
import numpy as np
```

```python
image = cv2.imread('eight.png', 0)
print(image.shape)
_, A = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
A_complement = cv2.bitwise_not(A)
X = np.zeros_like(A)
X[0, 0] = 255
kernel = cv2.getStructuringElement(cv2.MORPH_CROSS, (3, 3))
while True:
    X_prev = X.copy()
    X = cv2.dilate(X, kernel)
    X = cv2.bitwise_and(X, A_complement)
    if np.array_equal(X, X_prev):
        break
out = cv2.bitwise_or(A, X)

cv2.imshow('Original', A)
cv2.imshow('Filled Holes', out)
cv2.waitKey(0)
```
