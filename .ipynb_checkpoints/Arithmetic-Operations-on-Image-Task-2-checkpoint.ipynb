```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
try:
    img_bgr = cv2.imread('panda.png')

    if img_bgr is None:
        raise FileNotFoundError("Image file not found or could not be read.")
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
except FileNotFoundError as e:
    print(f"Error: {e}")
    print("Please ensure the image is in the correct directory if running locally.")

M_add = np.ones(img_bgr.shape, dtype="uint8") * 50

added_img_bgr = cv2.add(img_bgr, M_add)
added_img_rgb = cv2.cvtColor(added_img_bgr, cv2.COLOR_BGR2RGB)

M_sub = np.ones(img_bgr.shape, dtype="uint8") * 55
subtracted_img_bgr = cv2.subtract(img_bgr, M_sub)
subtracted_img_rgb = cv2.cvtColor(subtracted_img_bgr, cv2.COLOR_BGR2RGB)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('Simple Arithmetic Operations on Image Pixels', fontsize=14)

axes[0].imshow(img_rgb)
axes[0].set_title('1. Original Image')
axes[0].axis('off')

axes[1].imshow(added_img_rgb)
axes[1].set_title('2. Brightened (Image + 50)')
axes[1].axis('off')

axes[2].imshow(subtracted_img_rgb)
axes[2].set_title('3. Darkened (Image - 55)')
axes[2].axis('off')

plt.tight_layout()
plt.show()
```
