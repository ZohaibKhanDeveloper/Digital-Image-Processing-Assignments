```python
import cv2
import numpy as np
```

```python
image = cv2.imread('Panda.png', 0)
_, binary = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
kernel = np.ones((3, 3), np.uint8)
eroded_image = cv2.erode(binary, kernel, iterations=1)
boundary = cv2.subtract(binary, eroded_image)
cv2.imshow('Original (A)', binary)
cv2.imshow('Eroded (A minus B)', eroded_image)
cv2.imshow('Boundary Extraction', boundary)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
