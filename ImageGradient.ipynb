```python
import cv2
import numpy as np
```

```python
image = cv2.imread('Panda.png', 0)
_, binary_img = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

kernel = np.ones((3, 3), np.uint8)
gradient = cv2.morphologyEx(binary_img, cv2.MORPH_GRADIENT, kernel)
cv2.imshow('Original Binary Image', binary_img)
cv2.imshow('Morphological Gradient (Outlines)', gradient)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
