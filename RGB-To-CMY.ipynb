```python
import cv2
import numpy as np
```

```python
def rgb_to_cmy(img,img_name):
    img = img.astype(np.float32) / 255.0
    cmy = 1 - img
    cmy = (cmy * 255).astype(np.uint8)
    cv2.imwrite(f'{img_name}.png', cmy)
```

```python
panda = cv2.imread("Panda.png")
```

```python
cv2.imshow("Panda Image RGB",panda)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
rgb_to_cmy(panda,"PandaCMY")
```

```python
panda_cmy = cv2.imread("PandaCMY.png")
```

```python
cv2.imshow("Panda Image RGB",panda_cmy)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
